```python
df['Last_Name'] = df['Last_Name'].str.replace(r'[^a-zA-Z]', '', regex=True)
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[^0-9]', '', regex=True)
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: f"{x[:3]}-{x[3:6]}-{x[6:10]}" if pd.notna(x) else x)
```
